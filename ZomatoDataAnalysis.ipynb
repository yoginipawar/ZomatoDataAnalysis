```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**EDA**

Rate column

```python
df['rate'].unique()
```

Removing "NEW", "-" and "/5" from rate coloumn

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

Filling Null Valuse in Rate column with mean

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

Dropping Null value

```python
df.dropna(inplace = True)
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

Listed in(city) and location, both are there, lets keep only one.

```python
df = df.drop(['listed_in(city)'], axis = 1)
```

```python
df.head()
```

```python
df['Cost2plates'].unique()
```

Removing ',' from Cost2Plates *Column*

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)

df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

Cleaning Rest Type Column

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

Making Rest Types less than 1000 in frequency as others

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

Cleaning Location Column

```python
location = df['location'].value_counts(ascending  = False)

location_lessthan300 = location[location<300]



def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

Cleaning Cuisines Column

```python
cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

**8) No. of Votes, Location Wise**: In which locations people are intresting to give feedback

```python
df4 = df[['location', 'votes']]
df4.drop_duplicates()
df5 = df4.groupby(['location'])['votes'].sum() #total votes across all types at a given location.
df5 = df5.to_frame()
df5 = df5.sort_values('votes', ascending=False)
df5.head()
```

```python
# Prepare the data
df4 = df[['location', 'votes']].drop_duplicates()
df5 = df4.groupby('location')['votes'].sum().reset_index()
df5 = df5.sort_values('votes', ascending=False)

# Select top N locations for the pie chart (optional, to avoid overcrowding)
top_n = 10  # You can adjust this number based on your data
df5_top = df5.head(top_n)

# Create a pie chart
plt.figure(figsize=(10, 10))
plt.pie(df5_top['votes'], labels=df5_top['location'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('bright'))

# Set the title
plt.title('Proportion of Votes by Location')

# Show the plot
plt.show()
```

**Insight**: This pie chart displays the proportion of total votes across the top 10 locations. It helps identify which locations have received the highest number of votes, highlighting the most popular areas.

**9) Visualizing Top Cuisines**
